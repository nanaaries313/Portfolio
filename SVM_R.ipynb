```r
library(e1071) # For the svm() function
library(mlbench) # For the glass data
```

Our familiar glass data can be obtained directly from the mlbench package

```r
data(Glass, package="mlbench")
str(Glass)
```

Glass data is more complicated data, so a method like SVM could be helpful

Split into test-train (resampling may be more appropriate for such small data, but that's not our focus today)

```r
n=nrow(Glass)
set.seed(1234)
test = sample(1:n, as.integer(n/3))
test.data = Glass[test,]
train.data = Glass[-test,]
```

Run the svm() function with default parameters and construct the confusion matrix

```r
svm.model = svm(Type ~ ., data = train.data)
CM = table(predict(svm.model,test.data),
           test.data$Type)
CM
sum(diag(CM))/sum (CM)
svm.model # What kernel is used by default? Options?
```

By default, SVM model uses the radial kernel. Cost is 1 (error or penalty, in case the data points are not in the right side of the hyperplane)

These are the two hyperparameters that we can change

```r
#Total data points
n
#In the training data set
n*2/3
```

120 is very high suport vectors numbers, so SVM must match it with very high dimensional space to get this numbers

```r
?svm
```

Try a linear kernel

```r
svm.model.linear = svm(Type ~ ., data = train.data,
                       kernel="linear")
CM.linear = table(predict(svm.model.linear,test.data),
           test.data$Type)
CM.linear
sum(diag(CM.linear))/sum (CM.linear)
```

Try a polynomial kernel

```r
svm.model.poly = svm(Type ~ ., data = train.data,
                       kernel="polynomial")
CM.poly = table(predict(svm.model.poly,test.data),
                  test.data$Type)
CM.poly
sum(diag(CM.poly))/sum (CM.poly)

svm.model.poly # What is the degree of the polynomial?
```

Try a simpler polynomial kernel

```r
svm.model.poly.2 = svm(Type ~ ., data = train.data,
                       kernel="polynomial",
                       degree = 2)
CM.poly.2 = table(predict(svm.model.poly.2,test.data),
                  test.data$Type)
CM.poly.2
```

 Try a more complex polynomial kernel

```r
svm.model.poly.4 = svm(Type ~ ., data = train.data,
                       kernel="polynomial",
                       degree = 4)
CM.poly.4 = table(predict(svm.model.poly.4,test.data),
                  test.data$Type)
CM.poly.4
```

Different kernel and parameters work well for different issues. We do want to experiement to find out the best model

 This is very interesting!

We've tried the default (radial basis), linear and polynomiall kernels. There is one left to try

```r
svm.model.sigmoid = svm(Type ~ ., data = train.data,
                       kernel="sigmoid")
CM.sigmoid = table(predict(svm.model.sigmoid,test.data),
                   test.data$Type)
CM.sigmoid
```

 What parameters can we try to tune?

```r
svm.model.sigmoid$coef0 # Added constant in the kernel
svm.model.sigmoid$gamma # Multiplier in the kernel
svm.model.sigmoid$cost  # Penalty for violating constraints

svm.model.sigmoid = svm(Type ~ ., data = train.data,
                        kernel="sigmoid",
                        coef0 = 1,
                        gamma = 1)
CM.sigmoid = table(predict(svm.model.sigmoid,test.data),
                   test.data$Type)
CM.sigmoid
```

Tuning has a large effect! You now have a huge amount of parameter tuning that you can do to find the best SVM

This is the first time that you are working with a method that will allow a great deal of tuning.

##**Why do we tune?**
For better model

Find the kernel that works the best. And then tune for that particular kernel.

Closer to the training data? low bias. high variance.

Better model = bias-variance trade-off

How do I know if I got this right?

```r
svm.model = svm(Type ~ ., data = train.data)
CM = table(predict(svm.model,test.data),
           test.data$Type)
CM
sum(diag(CM))/sum (CM)

#what if we look at the training error?
CM.train = table(predict(svm.model,train.data),
           train.data$Type)
CM.train
sum(diag(CM.train))/sum (CM.train)
```

Boosted tree or deep learning are go-to methods for difficult datasets

```r
svm.model = svm(Type ~ ., data = train.data,
                cost = 0.01)
CM = table(predict(svm.model,test.data),
           test.data$Type)
CM
sum(diag(CM))/sum (CM)

#what if we look at the training error?
CM.train = table(predict(svm.model,train.data),
           train.data$Type)
CM.train
sum(diag(CM.train))/sum (CM.train)
```

This is generalizing as much as we can. This is exactly the same thing as KNN with high enough K, we predict the majority class
